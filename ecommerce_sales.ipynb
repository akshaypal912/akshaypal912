```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
df = data.sort_values('Order Date')
```

```python
df.set_index('Order Date', inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Sales'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
monthly_sales = df['Sales'].resample('ME').sum()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(monthly_sales)
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

Hence the p-vlaue is less than the 0.5 hence it is stationary ....


```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(monthly_sales, order=(1,1,1))
model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=6)
print(forecast)
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_sales, label="Actual Sales")
plt.plot(forecast, label="Forecasted Sales", color='red')
plt.legend()
plt.show()
```

```python
# The graph shows a clear seasonality in sales, with peaks typically occurring towards the end of each year (November-December).
# There is also an overall upward trend in sales over the years observed (2014-2017).
# The forecasted sales for the first half of 2018 show a decline after the year-end peak, which is consistent with the observed seasonal pattern.
```

```python
train = monthly_sales[:-12]   # last 12 months test
test  = monthly_sales[-12:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    train,
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

model_fit = model.fit()
print(model_fit.summary())
```

```python
forecast_6 = model_fit.forecast(steps=6)
```

```python
forecast_12 = model_fit.forecast(steps=12)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(monthly_sales, label='Actual Sales')
plt.plot(forecast_12, label='12-Month Forecast', color='red')
plt.legend()
plt.title("Sales Forecast (Next 12 Months)")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error

pred = model_fit.predict(start=test.index[0], end=test.index[-1])
mape = mean_absolute_percentage_error(test, pred)

print("MAPE:", round(mape*100, 2), "%")
```

```python
# The 'Sales Forecast (Next 12 Months)' graph compares the actual sales with the 12-month forecast generated by the SARIMAX model.
# It shows that the model generally captures the seasonal patterns, such as the year-end sales peaks.
# However, the Mean Absolute Percentage Error (MAPE) of 39.97% indicates that while the trend is captured, the precise sales figures in the forecast deviate significantly from the actual values, suggesting areas for potential model refinement or that sales are influenced by other unmodeled factors.
```
